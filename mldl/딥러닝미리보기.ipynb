### 에포크의 개념

> 데이터셋을 반복 학습

머신러닝 모델은 데이터를 여러 번 반복적으로 학습하는데, 에포크는 전체 데이터셋이 한 번 학습에 사용된 횟수를 나타냅니다.

예를 들어, 1 에포크는 데이터셋 전체가 한 번 모델에 입력되어 학습이 이루어졌다는 것을 의미합니다.

> 여러 번의 에포크

실전에서는 보통 하나의 에포크로는 모델이 충분히 학습되지 않기 때문에, 모델이 수십, 수백 번의 에포크 동안 학습을 반복합니다. 이를 통해 모델의 성능이 점진적으로 개선됩니다.

> 미니 배치 학습과의 관계

일반적으로 데이터셋이 크기 때문에, 한 번의 에포크에서 데이터셋 전체를 한꺼번에 모델에 입력하기보다는 데이터를 여러 개의 미니 배치(mini-batch)로 나누어 학습합니다.

각 미니 배치로 학습한 후 모델의 가중치가 업데이트되고, 모든 미니 배치가 처리되면 한 번의 에포크가 완료됩니다.

### 미니 배치(Mini-batch)

딥러닝 모델을 훈련할 때 전체 데이터셋을 여러 개의 작은 그룹으로 나누어 학습하는 방법입니다.
 
미니 배치를 사용하면 메모리 사용을 최적화하고 학습 효율을 높일 수 있습니다.
