### 예측 모델링 기법인 분류와 회귀의 구분 방법

>예측하려는 값의 형태 확인

이산적 범주(예: 클래스 레이블)라면 분류 문제입니다.

연속적인 실수 값이라면 회귀 문제입니다.

>목표와 목표 값의 특성 이해

데이터의 목표가 '그룹화'나 '카테고리화'라면 분류입니다. 예를 들어, "이 사진에 있는 동물은 무엇인가요?"는 분류 문제입니다.

데이터의 목표가 '양적 예측'이라면 회귀입니다. 예를 들어, "이 집의 가격은 얼마인가요?"는 회귀 문제입니다.

>평가 지표 확인

모델 성능을 평가할 때 정확도(accuracy), 정밀도(precision), 재현율(recall) 등을 사용한다면 분류입니다.

평균 제곱 오차(MSE), 평균 절대 오차(MAE) 등을 사용한다면 회귀입니다.

>>예시

분류 예시: 고객이 구매할 상품이 '전자제품', '의류', '식품' 중 어떤 것인지 예측하는 경우.

회귀 예시: 특정 집의 크기, 위치, 방 수를 기반으로 주택의 가격을 예측하는 경우.
