### Minkowski 거리

다양한 거리 계산 방법을 일반화한 수학적 거리 측정 방식입니다

이는 주로 머신러닝의 거리 기반 알고리즘(예: K-최근접 이웃 알고리즘, K-평균 클러스터링)에서 사용됩니다

### KNN(K-Nearest Neighbors, K-최근접 이웃)

매우 직관적이고 인기 있는 지도 학습 알고리즘

분류와 회귀 문제를 해결할 수 있으며, 주어진 데이터 포인트와 가장 가까운 K개의 이웃을 기반으로 예측을 수행합니다

```python
kn.kneighbors()
```

```python
len(kn.kneighbors())
```

```python
kn.score(fish_data,fish_target)
# score 모델의 성능을 평가한다
# 모델이 얼마나 잘 작동하는지 성능 점수를 반환한다

# 1.0인 경우 테스트 데이터의 모든 샘플을 정확히 분류했음을 의미한다
```

```python
kn.predict([[30,600]])
# knn 모델을 사용하여 새로운 데이터 포인트에 대한 예측을 수행한다

# 1이 나오면 1이라는 특정 클래스에 속한다는 뜻이다
```

```python
kn.predict([[13,50]])
```

### 리스트를 넘파이 배열로 변환하는 이유

>성능

리스트: Python 리스트는 일반적으로 느리며, 특히 대규모 데이터셋의 수치 연산에서 성능이 좋지 않습니다. 리스트는 동적으로 크기가 조절되며, 다양한 타입의 요소를 가질 수 있어서 유연하지만, 이로 인해 성능이 저하될 수 있습니다.

NumPy 배열: NumPy 배열은 메모리와 성능 측면에서 최적화되어 있습니다. 배열은 고정 크기이며, 같은 데이터 타입을 가진 요소들로만 구성됩니다. 이는 메모리 사용과 연산 속도를 대폭 향상시킵니다.

>수치 연산

리스트: 수치 연산을 할 때 리스트는 직접적인 지원을 제공하지 않으며, 연산을 위해 반복문을 사용해야 합니다.

NumPy 배열: NumPy 배열은 벡터화된 연산을 지원하여 빠르고 효율적으로 수치 연산을 수행할 수 있습니다. 예를 들어, 배열의 모든 요소에 대해 동일한 연산을 수행하는 것이 매우 빠릅니다.

>수학적 함수와 연산

리스트: 기본적인 수학 연산(덧셈, 뺄셈 등)을 직접 구현하거나 외부 라이브러리를 사용해야 합니다.

NumPy 배열: NumPy는 수학적 함수와 연산을 내장하고 있으며, 배열 전체에 대해 수학적 연산을 쉽게 적용할 수 있습니다. 예를 들어, 배열의 평균, 표준편차, 합계 등을 쉽게 계산할 수 있습니다.

>다차원 배열

리스트: 다차원 리스트를 만들 수 있지만, 다차원 배열에 대한 연산과 조작이 복잡하고 비효율적일 수 있습니다.

NumPy 배열: 다차원 배열을 자연스럽게 지원하며, 행렬 연산 및 복잡한 수치 계산을 쉽게 수행할 수 있습니다.

>인덱싱과 슬라이싱

리스트: 인덱싱과 슬라이싱은 가능하지만, 다차원 배열을 다루는 데는 제한이 있습니다.

NumPy 배열: NumPy 배열은 고급 인덱싱 및 슬라이싱을 지원하여 복잡한 데이터 조작을 쉽게 수행할 수 있습니다. 예를 들어, 특정 조건에 맞는 요소를 필터링하거나, 배열의 특정 부분을 선택할 수 있습니다.
