```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import pandas as pd
```

```python
pd.read_csv('./data/sonar3.csv',header=None)
# 광석 여부 판별
```

```python
df=pd.read_csv('./data/sonar3.csv',header=None)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#기본값 훈련셋, 테스트셋
#0.75:0.25
```

```python
X=df.iloc[:,:60].values
y=df.iloc[:,60].values
```

```python
208*0.7
```

```python
train_test_split(X,y,test_size=0.3,shuffle=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,shuffle=True)
```

```python
X_train.shape
```

```python
208-145
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model=Sequential()
model.add(Dense(36,input_dim=60,activation='relu'))
model.add(Dense(18,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
model.fit(X_train,y_train,epochs=200)
```

```python
#실제 모델의 정확도는 테스트용으로 확인해야함
```

```python
model.evaluate(X_test,y_test)
```

```python
score=model.evaluate(X_test,y_test)
```

```python
#테스트 셋에서의 정확도를 가지고 판단해야함 79.3%
```

```python
model2=Sequential()
model2.add(Dense(24,input_dim=60,activation='relu'))
model2.add(Dense(10,activation='relu')) #덴스가 보통 입력층의 반
model2.add(Dense(1,activation='sigmoid'))
model2.compile(loss='binary_crossentropy',
              optimizer='adam',metrics=['accuracy'])
model2.fit(X_train,y_train,epochs=200,batch_size=10)
# 정확도가 100인 경우 과적합일 수 있다
```

```python
score2=model2.evaluate(X_test,y_test)
```
