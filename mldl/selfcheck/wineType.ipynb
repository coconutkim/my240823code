```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
# 데이터셋을 무작위로 훈련용과 테스트용 집합으로 나누는데 사용된다

# 훈련 데이터를 사용해 모델을 학습시키고,
# 테스트 데이터를 통해 학습된 모델의 일반화 성능을 평가한다

import pandas as pd
```

```python
!git clone https://github.com/taehojo/data.git
```

```python
df=pd.read_csv('./data/wine.csv',header=None)
# 헤더가 없으므로 추가한다
df
```

```python
X=df.iloc[:,0:12] #와인의 속성, 즉 0부터 11까지
y=df.iloc[:,12] #와인의 분류, 1이냐 0이냐
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,
                                                 test_size=0.2,
                                                 shuffle=True)
# 전체 데이터의 20퍼센트는 테스트, 80퍼센트는 훈련 세트로 사용한다
# 데이터를 나누기 전에 무작위로 섞겠다
```

```python
model=Sequential()
```

```python
model.add(Dense(30,input_dim=12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
# 은닉층에서는 주로 렐루, 출력층에서는 시그모이드나 다른 함수
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',
              metrics=['accuracy'])
# loss 모델이 학습하는 동안 얼마나 잘못 예측하는지 측정하는 함수
# 이진 분류 문제이기에 바이너리 크로스엔트로피를 활용한다

# optimizer 모델의 가중치를 조정하여 손실 함수를 최소화한다

# metrics 모델의 성능을 평가하는데 사용되는 기준
```

```python
history=model.fit(X_train,y_train,epochs=50,batch_size=500,
                  validation_split=0.25)
# fit 주어진 훈련 데이터를 사용하여 모델을 학습시킨다
# x 입력 데이터
# y 예측하려고 하는 정답
# 50번의 반복 학습
# 500개의 모델이 한번에 처리하는 데이터 샘플의 수
# 훈련 데이터의 25퍼센트를 검증 데이터로 사용한다
```

```python
score=model.evaluate(X_test,y_test) #모델의 성능을 평가한다
print('test accuracy:',score[1]) #테스트 데이터에 대한 모델의 정확도 출력
```
